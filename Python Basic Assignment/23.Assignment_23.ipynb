#### 1. What is the result of the code, and why?
`>>> def func(a, b=6, c=8):`   
`print(a, b, c)`   
`>>> func(1, 2)`   
**Ans:** The result of the above code is `1 2 8`. its because the function uses the default of c ie 8 which  is provided at the time of declaration

```python
def func(a,b=6,c=8):
    print(a,b,c)
func(1,2)
```

#### 2. What is the result of this code, and why?
`>>> def func(a, b, c=5):`  
`print(a, b, c)`  
`>>> func(1, c=3, b=2)`     
**Ans:** The result of the above code is `1 2 3`. it is because the function will use default values only when a value for a argument is not provided and if argument name is mentioned while doing a function call, the order of arguments is also ignored by the python interpreter

```python
def func(a,b,c=5):
    print(a,b,c)
func(1,c=3,b=2)
```

#### 3. How about this code: what is its result, and why?
`>>> def func(a, *pargs):`   
`print(a, pargs)`   
`>>> func(1, 2, 3)`   

```python
def func(a, *pargs):
    print(a,pargs)
func(1,2,3)
```

#### 4. What does this code print, and why?
`>>> def func(a, **kargs):`   
`print(a, kargs)`   
`>>> func(a=1, c=3, b=2)`   

```python
def func(a,**kargs):
    print(a,kargs)
func(a=1,c=3,b=2)
```
